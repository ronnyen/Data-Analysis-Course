```python
import requests 
import pymongo
import random
import pandas as pd

API_KEYS = ['']
uri = 'mongodb://ronny:12345@ds231228.mlab.com:31228/videos' 
```

```python
df = pd.read_csv('data/channels_id.csv')
len(df) # 1650
```

```python
# list of ids to call the api 
channel_ids = df.loc[1601:,'channelId']
```

```python
# list for all the api responses
channels_json = []

# call the api
for channel in channel_ids:
    k = random.randint(0,3)
    API_KEY = API_KEYS[k]
    url = "https://www.googleapis.com/youtube/v3/channels?part=contentOwnerDetails,localizations,snippet,statistics,topicDetails&id="+channel+"&key="+API_KEY
    r = requests.get(url)
    get_json = r.json()
    channels_json.append(get_json)

# insert videos_json to mongodb
client = pymongo.MongoClient(uri)
db = client.get_database()
collection = db['channels']
collection.insert_many(channels_json)
client.close()
```

```python
df = pd.read_csv('data/channels_id.csv', index_col=0)
```

```python
# mongodb to csv
# tried to take id from video.csv file and add more data from mongodb
client = pymongo.MongoClient(uri)
db = client.get_default_database()
ids = df.index[1526:]
for id_ in ids:
    channel = db.get_collection('channels').find({'items.id': id_})
    for c in channel:
        try:
            df.loc[id_, 'country'] = c['items'][0]['snippet']['country']
        except:
            df.loc[id_, 'country'] = None
        df.loc[id_, 'description'] = c['items'][0]['snippet']['description']  
        df.loc[id_, 'title'] = c['items'][0]['snippet']['title']
        try:
            df.loc[id_, 'publishedAt'] = c['items'][0]['snippet']['publishedAt']
        except:
            df.loc[id_, 'publishedAt'] = None   
        df.loc[id_, 'thumbnails'] = c['items'][0]['snippet']['thumbnails']['default']['url']
        df.loc[id_, 'commentCount'] = c['items'][0]['statistics']['commentCount']
        df.loc[id_, 'subscriberCount'] = c['items'][0]['statistics']['subscriberCount']
        df.loc[id_, 'videoCount'] = c['items'][0]['statistics']['videoCount']
        df.loc[id_, 'viewCount'] = c['items'][0]['statistics']['viewCount']
        try:
            df.loc[id_, 'topicIds'] = str(c['items'][0]['topicDetails']['topicIds'])
        except:
            df.loc[id_, 'topicIds'] = None
            
df[1526:].to_csv('data/channels.csv',  mode='a', header=None)
df
```
