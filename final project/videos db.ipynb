```python
import requests 
import pymongo
import random
import pandas as pd

API_KEYS = []
uri = 'mongodb://ronny:12345@ds231228.mlab.com:31228/videos' 
```

```python
# list of 3685 videos id, videos that appeared on the trending list from 14/11/17 to 7/2/18 
df = pd.read_csv('data/videos_id.csv')
len(df) # 3685
# # split the list between us
# # efrat
# df = df[:1000]
# # yaara
# df = df[1000:2000].reset_index()
# # ronny
# df = df[2000:].reset_index()
```

```python
# list of ids to call the api 
video_ids = df.loc[:,'video_id']
```

```python
# list for all the api responses
videos_json = []

# call the api
for video in video_ids:
    k = random.randint(0,3)
    API_KEY = API_KEYS[k]
    url = "https://www.googleapis.com/youtube/v3/videos?part=contentDetails,recordingDetails,topicDetails,snippet,statistics&id="+video+"&key="+API_KEY
    r = requests.get(url)
    get_json = r.json()
    videos_json.append(get_json)

# insert videos_json to mongodb
client = pymongo.MongoClient(uri)
db = client.get_database()
collection = db['yourname']
collection.insert_many(videos_json)
client.close()
```

```python
# combine collections
# https://stackoverflow.com/questions/11554762/how-to-copy-a-collection-from-one-database-to-another-in-mongodb?rq=1
client = pymongo.MongoClient(uri)
db = client.get_database()
cursor = db["ronny"].find()
db["videos"].insert_many(cursor)
client.close()
```

```python
df = pd.read_csv('data/videos_id.csv', index_col=0)
```

```python
# mongodb to csv
# tried to take id from video.csv file and add more data from mongodb
client = pymongo.MongoClient(uri)
db = client.get_default_database()
t = df.index[3083:]
for i in t:
    videos = db.get_collection('videos').find({'items.id': i})
    for video in videos:
        df.loc[i, 'channelId'] = video['items'][0]['snippet']['channelId']
        df.loc[i, 'liveBroadcastContent'] = video['items'][0]['snippet']['liveBroadcastContent']  
        df.loc[i, 'definition'] = video['items'][0]['contentDetails']['definition']
        df.loc[i, 'dimension'] = video['items'][0]['contentDetails']['dimension']
        df.loc[i, 'duration'] = video['items'][0]['contentDetails']['duration']
        try:
            df.loc[i, 'relevantTopicIds'] = str(video['items'][0]['topicDetails']['relevantTopicIds'])
        except:
            df.loc[i, 'relevantTopicIds'] = None
            
df[3083:].to_csv('data/videos.csv',  mode='a', header=None)
df
```
