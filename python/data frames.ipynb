## The .apply() Method with Row Values

```python
# custom function
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "I have no clue"
# apply - moving across the columns    
bond.apply(good_movie, axis="columns")
```

# .groupby() 

```python
fortune = pd.read_csv("data/fortune1000.csv", index_col = "Rank")
fortune.head(3)
```

```python
# groupby create a bunch of df, df for each group
sectors = fortune.groupby("Sector")
sectors
```

```python
# number of groups
len(sectors)
```

```python
sectors.size() # number of members in a group
```

```python
# extract the first/last row of every group
sectors.first()
```

```python
sectors.groups # dict of the index label of each row in the group
```

## .get_group()

```python
sectors.get_group("Apparel").sum()
```

```python
sectors.max()
sectors.min()
sectors.sum()
sectors.mean().head(3)
```

## Grouping by Multiple Columns

```python
fortune = pd.read_csv("data/fortune1000.csv", index_col = "Rank")
sectors = fortune.groupby(["Sector", "Industry"])
fortune.head(3)
```

```python
sectors.size()
sectors.sum()
sectors["Revenue"].sum()
sectors["Employees"].mean().head(3)
```

```python
sectors.get_group(('Apparel', 'Apparel'))
```

## .agg()

```python
fortune = pd.read_csv("data/fortune1000.csv", index_col = "Rank")
sectors = fortune.groupby("Sector")
fortune.head(3)
```

```python
sectors.agg({'Revenue' : ['mean', 'sum'],
            'Profits' : 'mean'}).head(3)
```

```python
sectors.agg(['size', 'sum', 'mean']).head(3)
```

## Iterating through Groups

```python
df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for sector, data in sectors:
    highest_revenue_company_in_group = data.nlargest(1, "Revenue")
    df = df.append(highest_revenue_company_in_group)
df.head(3)
```
